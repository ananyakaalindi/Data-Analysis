```python
#Merging two dataframes

dict1={'Objects':['pen','pencil','bag','mug'],'Price':[20,10,200,70]}
df1=pd.DataFrame(dict1)
dict2={'Objects':['cup','pencil','pen'],'Color':['Red','Ash','Black'],'Price':[50,15,20]}
df2=pd.DataFrame(dict2)
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='Price') #criteria defined for merging
```

```python
pd.merge(df1,df2,on='Price',how='left')
```

```python
pd.merge(df1,df2,on='Price',how='right')
```

```python
pd.merge(df1,df2,on='Price',how='outer')
```

```python
pd.merge(df1,df2,on='Price',how='inner')
```

```python
#using indexes as keys for merging

pd.merge(df1,df2,right_index=True,left_index=True)
```

```python
#join function can be used to do the merging by indexes

#It can combine many dataframes having same indexes but no columns overlapping.

df2.columns=['Obj','Color','Rate']
df1.join(df2)
```

```python
frame1=pd.DataFrame({'a':[1,5,8,7,6,5],'b':[7,5,9,np.nan,4,np.nan],'c':[5,9,3,2,6,5]})
frame2=pd.DataFrame({'a':[1,8,2,7,5,np.nan],'b':[7,5,np.nan,9,np.nan,4],'c':[5,9,3,2,6,5]},index=[6,7,8,9,10,11])
pd.concat([frame1,frame2])
```

```python
pd.concat([frame1,frame2],axis=1)
```

```python
dict1={'Color':['green','red','blue','green','blue','black'],'Value':[3,7.8,4,6,9,2.2]}
df=pd.DataFrame(dict1)
df.groupby('Color').sum()
```

```python
#Pivot Tables
data=pd.read_csv('iris.csv')
df=pd.DataFrame(data)
df1=df.head() 
table=df1.pivot_table(index='sepallength')
table
```

```python
table=df1.pivot_table(index='sepallength',columns='petallength',aggfunc=np.sum)
table
```
