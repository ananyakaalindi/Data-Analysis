```python
import pandas as pd

#defining a dataframe
data={'Name':['Ram','Sanjay','Hima','Laya','Meena','Rishi'],'Age':[12,45,13,23,35,54],'Graduation':[False,True,False,True,True,True]}
df=pd.DataFrame(data)
df
```

```python
#extracting only required columns
df2=pd.DataFrame(data,columns=['Name','Graduation'])
df2
```

```python
#changing the index values
df=pd.DataFrame(data,index=['a','b','c','d','e','f'])
df
```

```python
#To know the names of columns or rows in a dataset
print(df.columns)
print(df.index)
```

```python
print(df['Age']) #to extract data of a specific column
print(df.loc['b']) #of a specific row
print(df.loc[['c','e']]) #of multiple rows
```

```python
print(df['a':'c'])
print(df['Age'][3])
```

```python
#Assigning values
df.index.name='id'
df.columns.name='Details'
df
```

```python
#Adding a new column
df['Height']=155
df
```

```python
df['Height']=[140,150,155,160,153,167]
df
```

```python
import numpy as np

ser=pd.Series(np.arange(6),index=['a','b','c','d','e','f'])
df['New']=ser #Adding a series as a column
df
```

```python
del df['New'] #deleting a column
df
```

```python
#When Nested Dict is passed as an argument to the DataFrame(), external keys will be treated as columns and internal keys as index labels

dict={'Sales':{2009:3000,2010:3500,2011:3200},'Profit':{2009:54,2010:66,2011:57}}
df3=pd.DataFrame(dict)
df3
```

```python
#While adding, when same labels are present in both they will be added

frame1=pd.DataFrame(np.arange(16).reshape((4,4)),index=['blue','yellow','green','red'],columns=['pen','pencil','bag','cup'])
frame2=pd.DataFrame(np.arange(11,27).reshape((4,4)),index=['yellow','black','red','blue'],columns=['pencil','paper','cup','pen'])
frame1+frame2
```

```python
frame1.mul(frame2) #add(),sub(),mul(),div() can be performed this way
```

```python
print(frame1)
frame1.sum()
```

```python
frame1.sort_index()
```
