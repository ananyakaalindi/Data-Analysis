# Download the iris data set and read into a pandas data frame. Extract the header and replace with a new header. Extract colums and rows. Extract pivot tables. Filter the data based on the labels. Store a pivot table as a pickle and retrieve it.  

```python
import numpy as np
import pandas as pd

df=pd.read_csv("iris - iris.csv") #reading the iris dataframe

cols=df.columns #extracting header
print(cols)

cols1=['slength','swidth','plength','pwidth','classes'] #replacing with a new header
df.columns=cols1
print(df)

a=df[['slength','plength']] #extracting specific columns
print(a)

b=df.loc[2] #extracting specific row
print(b)

```

```python
f=df.pivot_table(index='slength',columns='plength') #created pivot table
print(f)

f.to_pickle('iris_pivot') #pivot table stored as pickle file
```

```python
unpickled_f=pd.read_pickle('iris_pivot')
unpickled_f
```
